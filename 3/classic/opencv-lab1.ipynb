```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import peak_local_max

img = cv2.cvtColor(cv2.imread('penguins.jpg'), cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img)
plt.show()

# local maximas
xy = peak_local_max(gray, min_distance=20,threshold_abs=100)
print(len(xy))
for (p1,p0) in xy:
    cv2.circle(img, (p0, p1), 5, (255,255,0), 2)
plt.imshow(img)
plt.show()
```
