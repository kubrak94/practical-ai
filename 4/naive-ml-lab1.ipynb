```python
import csv
import matplotlib.pyplot as plt

labels = []
incomes = []
days = []
X = []
with open('income.csv') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        label = row[0]
        day = float(row[1])
        rub_dollar = float(row[2])
        income = float(row[3])

        X.append([day, rub_dollar])
        labels.append(label)
        incomes.append(income)
        days.append(day)

plt.figure(figsize=(20,5))
plt.xticks(days, labels, rotation='vertical')
plt.plot(days, incomes)
plt.show()
```
